```python
from astral import LocationInfo
from astral.sun import elevation, azimuth, zenith
from datetime import datetime
from datetime import timedelta
import pytz, matplotlib.pyplot as plt
import numpy as np
```

```python
# Define location (City, Region, Country, Latitude, Longitude)
location = LocationInfo('St. Gallen', 'Switzerland', 'Europe/Zurich', 47.424492554512014, 9.376722938498643)

timestamp = datetime(2024, 1, 1, 3, 0, 0, tzinfo=pytz.timezone("Europe/Zurich"))

times = []
zen = []
zensine = []
zencos = []

for i in range(365*24*6):
    timestamp = timestamp + timedelta(minutes=10)
    zen.append(elevation(location.observer, timestamp))
    zensine.append(np.sin(np.radians(zen[-1])))
    zencos.append(np.cos(np.radians(zen[-1])))

    times.append(timestamp)
```

```python
plt.figure(figsize=(24, 6))
plt.plot(times, zen, label="Zenith")
plt.legend()
```

```python
plt.figure(figsize=(24, 6))
# plt.plot(times, zen, label="Zenith")
n = 365*24*6
plt.plot(times[:n], zensine[:n], label="sin(Zenith)")
plt.plot(times[:n], zencos[:n], label="cos(Zenith)")
plt.legend()
plt.show()
```
