```python
import pandas as pd
import numpy as np
from geographiclib.geodesic import Geodesic
```

```python
df = pd.read_csv('../../data/webcams-2024-10-14-v2.csv', sep=';', encoding='utf-8')
```

```python
stgallen = (47.42435370465393, 9.37670658305518)
geod = Geodesic.WGS84
```

```python
df["apx_dir_to_stg"] = df.apply(lambda row: geod.Inverse(row["lat"], row["lon"], stgallen[0], stgallen[1])["azi1"], axis=1)
```

```python
df["apx_dir_fr_stg"] = df.apply(lambda row: geod.Inverse(stgallen[0], stgallen[1], row["lat"], row["lon"])["azi1"], axis=1)
```
